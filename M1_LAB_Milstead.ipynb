```python
#M1T1_Milstead
#8/22/2018

import math

"""
Author: Mallory Milstead
Program uses the pythagorean theorem to calculate the hypotenuse of a right triangle and displays that information to the user
"""

def main():
    """
    Calculate either the hypotenuse or length of sides of a right triangle using the Pythagorean Theorem.
    
    Input: Choose menu option 1 or 2

    Option 1 Input: Length and width of box (right triangle)
             Output: The largest pole that could fit in the box (hypotenuse)
    Option 2 Input: Length of the pole (hypotenuse)
             Output: The smallest box that the pole can fit into (in whole integers)
    
    """
    #Initalize variable to make the while loop run once
    runAgain = "Y"
    while runAgain == "Y":
      userChoice = getUserChoice()
      if userChoice == "1":
        maxPoleLength()
      elif userChoice == "2":
        getBoxSize()
      else:
        while userChoice != "1" or userChoice != "2":
          print("That input is not valid, please enter 1 or 2 \n \n")
          userChoice = getUserChoice()
          if userChoice == "1":
            maxPoleLength()
            break
          elif userChoice == "2":
            getBoxSize()
            break
          
      runAgain = input("Would you like to run the program again? Y or N: " + "\n").upper() #Get user input to determine if while loop will run again
      
    print("Goodbye")

def getUserChoice():
	"""Ask the user to choose option "1" or "2" to run the program"""
	userChoice =input('If you know the size of the box and want to calculate the longest pole that can fit inside, enter "1" \
	 \n \nIf you know the length of the pole and want to calculate an appropriate box size, enter "2" \n \n')
	return userChoice

def getHypotenuse(width = .01, length = .01):
   """Allow the user to input the width and length of a right angle triangle and return the hypotenuse"""
   width = int(input("Enter the width of the box: "))
   length = int(input("Enter the length of the box: "))
   hypotenuse = math.sqrt(width**2 + length**2)
   return hypotenuse 

def maxPoleLength():
    """Uses the pythagorem theorem to determine if the rodLength is less than or equal to the
    hypotenuse of the box"""
    hypotenuse = getHypotenuse()
    return print("The largest pole that could fit in the box is", format(hypotenuse, ".2f") + "\n")

def getBoxSize():
  """Allow the user to input the length of the fishing pole and determine the smallest box it could fit into"""

  rodLength = int(input("How long is the fishing pole? "))

  #Initialize variables for the sides of the right triangle
  width = 1
  length = 1

  hypotenuse = math.sqrt(width**2 + length**2)
  while hypotenuse < rodLength: #While the hypotenuse is less than the rodLength
    width += 1 #Increment width
    hypotenuse = math.sqrt(width**2 + length**2) #Calculate hypotenuse
    if hypotenuse < rodLength: #If it is still less than the rodLength
      length += 1 #Increment the length
      hypotenuse = math.sqrt(width**2 + length**2) #Calculate hypotenuse and return the program to the while loop
  return print("The box must be at least", format(width, ".2f"), "by", format(length, ".2f") + "\n" +"\n" + ("-" * 50) + "\n" + \
               "width = " + str(width) + "\t" + "length = " + str(length) + "\t" + "hypotenuse = " + format(hypotenuse, ".2f") + "\n")



     

if __name__ == "__main__":
    main()
```
