```python
# CSC221
# M2HW_Milstead
# Goal: Gold

"""
Author: Mallory Milstead
Allows user to choose from menu options including searching and sorting lists and dictionaries
"""

import csv
from itertools import islice
import pickle

def main():

    """
    Allows the user to choose an option from the menu.

    Input: The menu choice that the user would like to run (an int)
    Output: Pokemon data specific to the menu choice
    """

    runAgain = "y"

    while runAgain == "y":
    
        chooseOption = str(input("Choose an option: \nTo list all pokemon, enter 1 \nTo search for a pokemon, enter 2 \nTo search for a pokemon in a Dictionary, enter 3\nTo Save Pokemon Data, enter 4\nTo Load Pokemon Data, enter 5\
                                    \nTo Display sorted Pokemon data, enter 6\n"))
        if chooseOption == "1":
            namesList = printNames('pokemon.csv')
    
            for name in namesList:
                print(name, end=", ")
            print("\n")


        if chooseOption == "2":
            outerList = getRecordsFromList('pokemon.csv')
            userInput = str(input("\nEnter the name (or partial name) of a pokemon to see it's stats: ").lower())

            #Create an empty list to hold the found items
            foundPoke = []

            
            for item in outerList:
                if userInput in item[1]:
                    foundPokemon = (item[0],item[1],item[2],item[3])
                    foundPoke.append(foundPokemon)
            for item in foundPoke:
                print('ID: {:10s}Name: {:22s}Height: {:15s}Weight: {:15s}'.format(item[0], item[1], item[2], item[3]))
                print("")
            if foundPoke == []:
                print("\nThat pokemon could not be found\n")
            
            

        if chooseOption == "3":

            dictList = getRecordsFromDict('pokemon.csv')
            userInput = str(input("\nEnter the name of a pokemon to see it's stats: ").lower())
            
            #Create an empty list to hold the found items
            foundPoke = []

            #If user input matches the value for key["Name"], add all of the items(by key) of that pokemon to a list
            for item in dictList:
                if userInput in item["Name"]:
                    foundPokemon = (item["ID"],item["Name"],item["Height"],item["Weight"])
                    foundPoke.append(foundPokemon)
            for item in foundPoke:
                print('ID: {:10s}Name: {:22s}Height: {:15s}Weight: {:15s}'.format(item[0], item[1], item[2], item[3]))
                print("")
            if foundPoke == []:
                print("That pokemon could not be found")

        if chooseOption == "4":
            outerList = getRecordsFromList('pokemon.csv')
            pickleData(outerList, "pickled_poke.txt")
            print("\nYour list has been pickled...\n")

        if chooseOption == "5":
            unpickledList = unpickleData(outerList, "pickled_poke.txt")
            print("\nYour list has been unpickled...\n")

        if chooseOption == "6":
            outerList = getRecordsFromList('pokemon.csv')
            pickleData(outerList, "pickled_poke.txt")
            unpickledList = unpickleData(outerList, "pickled_poke.txt")
            unpickledList.sort(key=lambda x: x[1])
            for item in unpickledList:
                print('ID: {:10s}Name: {:22s}Height: {:15s}Weight: {:15s}'.format(item[0], item[1], item[2], item[3]))
            

        runAgain = str(input("Would you like to run the program again? y/n ").lower())
    


def printNames(filename):
    """
    Returns a list of all 151 Pokemon names

    Input: filename (string)
    Output: a list of names
    """
    
    inputFile = open(filename, "r")
    csv_reader = csv.reader(inputFile)

    list_of_pokemon = []
    
    for line in islice(csv_reader, 1, 152):
        list_of_pokemon.append(line[1])

    inputFile.close()
    return list_of_pokemon


def getRecordsFromList(filename):
    """
    Returns a list of lists. The inner lists are each pokemon's int ID, identifier(name), height and weight
    Input: filename (string)
    Output: a list of lists
    """
    
    inputFile = open(filename, "r")
    csv_reader = csv.reader(inputFile)

    outerList = []

    for line in islice(csv_reader, 1, 152):
        pokeID = line[0]
        identifier = line[1]
        height = line[3]
        weight = line[4]
        
        innerList = [pokeID, identifier, height, weight]
        outerList.append(innerList)

        
    return outerList

def getRecordsFromDict(filename):
    """
    Returns a list of dictionaries with the pokemon's int ID, name, height and weight as values
    Input: filename (string)
    Output: a list of dictionary items
    """
    
    inputFile = open(filename, "r")
    csv_reader = csv.reader(inputFile)

    dictList = []

    for line in islice(csv_reader, 1, 152):
        pokeID = line[0]
        identifier = line[1]
        height = line[3]
        weight = line[4]
        
        pokeDict = {"ID" : pokeID, "Name": identifier, "Height": height, "Weight": weight}
        dictList.append(pokeDict)

        
    return dictList

def pickleData(listname, filename):
    """
    Pickles a list

    Input: the name of a list to be pickled, the filename for the pickled list
    Output: a pickled list with the given filename
    """
    
    writeFile = open(filename, "wb")
    for item in listname:
        pickle.dump(item, writeFile)
    writeFile.close()

def unpickleData(listname,filename):
    """
    Unpickles a list

    Input: the name to give to the unpickled list, the filename to unpickle
    Output: a pickled list with the given listname
    """
        
    listname = []
    readFile = open(filename, "rb")
    while True:
        try:
            item = pickle.load(readFile)
            listname.append(item)
        except EOFError:
            readFile.close()
            break
    return listname
    


if __name__ == "__main__":
    main()
```
